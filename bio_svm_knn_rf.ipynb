```python
df.describe()
```

```python
# 'state' sütununu bırakalım
df = df.drop(columns=['state'])
```

```python
# 'state_numeric' sütununu etiket olarak kullanalım ve etiketlerin sayısını görelim
labels = df['state_numeric']
print(labels.value_counts())
```

Yukarıda label olarak kullanılan state_numeric için value_counts() yapılarak hasta ve sağlıklı olan toplam örnek sayısı bulunmuştur.

```python
from matplotlib import pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='state_numeric', discrete=True)
plt.title('Distribution of Labels')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.xticks([0, 1])  # X ekseninde sadece 0 ve 1 değerlerini gösterelim
plt.show()
```

```python
# label olarak kullanılan state_numeric drop edilir
df = df.drop(columns=['state_numeric'])
```

```python
# Veri setinin %80'i train ve %20'si test verisi olacak şekilde ayrılır.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, labels, test_size=0.2, random_state=0, stratify = labels)
# stratify=labels parametresi, sınıf dağılımlarının eğitim ve test kümelerinde dengeli olmasını sağlar.
```

```python
print("X_train:", X_train.shape, "X_test", X_test.shape)
print("y_train:", y_train.shape, "y_test", y_test.shape)
```

Veri setinde toplamda 156 örnek olduğunu söylemiştik.

Bunun %80'i olan 124 adet örnek train verisi olarak seçilir. Kalan 32 adet örnek ise test verisi olarak kullanılmak üzere ayrılır.

**SVM**

```python
from sklearn import svm

clf = svm.SVC()

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Macro Precision:",metrics.precision_score(y_test, y_pred,average='macro'))
print("Macro Recall:",metrics.recall_score(y_test, y_pred,average='macro'))
```

```python
target_names = ['Tumor', 'Healthy']
```

```python
from sklearn.metrics import classification_report
# Sınıflandırma raporunu yazdıralım
print("SVM Classification Report:")
print(classification_report(y_test, y_pred, target_names=target_names))
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier

# KNN modelini oluşturalım
knn = KNeighborsClassifier(n_neighbors=5)

# Modeli eğitim veri setiyle eğitelim
knn.fit(X_train, y_train)

# Test veri seti üzerinde tahmin yapalım
y_pred_knn = knn.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_knn))
print("Macro Precision:",metrics.precision_score(y_test, y_pred_knn,average='macro'))
print("Macro Recall:",metrics.recall_score(y_test, y_pred_knn,average='macro'))
```

```python
print("KNN Classification Report:")
print(classification_report(y_test, y_pred_knn, target_names=target_names))
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest modelini oluşturalım
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Modeli eğitim veri setiyle eğitelim
rf.fit(X_train, y_train)

# Test veri seti üzerinde tahmin yapalım
y_pred_rf = rf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_rf))
print("Macro Precision:",metrics.precision_score(y_test, y_pred_rf,average='macro'))
print("Macro Recall:",metrics.recall_score(y_test, y_pred_rf,average='macro'))
```

```python
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf, target_names=target_names))
```

**En Başarılı Modelin Tespit Edilmesi**

```python
from sklearn.metrics import f1_score

# Modellerin F1-skorlarını hesaplayalım
svm_f1_score = f1_score(y_test, y_pred)
knn_f1_score = f1_score(y_test, y_pred_knn)
rf_f1_score = f1_score(y_test, y_pred_rf)

# F1-skorlarını bir sözlükte saklayalım
model_f1_scores = {
    "SVM": svm_f1_score,
    "KNN": knn_f1_score,
    "RF": rf_f1_score
}

# Kullanılan üç modelin F1 skorlarını bastıralım
for model, f1_score in model_f1_scores.items():
    print(f"{model} F1 Skoru: {f1_score:.4f}")

# En iyi F1-skora sahip modeli bulalım
best_model_name = max(model_f1_scores, key=model_f1_scores.get)
best_model_score = model_f1_scores[best_model_name]

# Sonucu ekrana yazdıralım
print(f"En başarılı model: {best_model_name}, başarı: {best_model_score}")
```

**K-Fold Cross Validation for SVM, KNN and RF**

Kullanılan üç farklı makine öğrenmesi modelinin performansını daha doğru tespit etmek adına 10 katlı çapraz doğrulama (10-fold cross-validation) yapılmıştır.

```python
from sklearn.model_selection import cross_val_score

# 10-Fold Cross Validation for SVM
from sklearn.svm import SVC
clf_svc = SVC() # clf: classifier
scores = cross_val_score(clf_svc, X_train, y_train, cv=10)
print("SVM: %0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))

# 10-Fold Cross Validation for KNN
from sklearn.neighbors import KNeighborsClassifier
clf_knn = KNeighborsClassifier(n_neighbors=5)
scores = cross_val_score(clf_knn, X_train, y_train, cv=10)
print("KNN: %0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))

# 10-Fold Cross Validation for Random Forest
from sklearn.ensemble import RandomForestClassifier
clf_rf = RandomForestClassifier(n_estimators=100, random_state=42)
scores = cross_val_score(clf_rf, X_train, y_train, cv=10)
print("Random Forest: %0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

**Hyperparameter Optimization**

SVM modeli kullanılmıştır.

```python
from sklearn.model_selection import GridSearchCV
# Hyperparameter Optimization
parameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}

svc = svm.SVC()

# Run the grid search
grid_obj = GridSearchCV(svc, parameters, cv=5)
grid_obj = grid_obj.fit(X_train, y_train)

# show best parameter configuration found for classifier
best_parameters = grid_obj.best_params_
print(best_parameters)

# Set the clf to the best combination of parameters
best_clf = grid_obj.best_estimator_

y_pred = best_clf.predict(X_test)

print(classification_report(y_test, y_pred, target_names=target_names))
```

**Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred, labels=best_clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_clf.classes_)
disp.plot()
plt.show()
```

**Model Performansını Etkileyen En Etkili Genler**
- En başarılı performansı veren bir diğer model olan Random Forest modeli kullanılmıştır.



```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# Random Forest modelini eğitelim
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Özellik önemliliğini alalım
feature_importances = rf.feature_importances_

# Özellik isimleri ile birlikte DataFrame'e dönüştürelim
features = pd.DataFrame({'Feature': X_train.columns, 'Importance': feature_importances})

# Önemliliğe göre sıralayalım
features = features.sort_values(by='Importance', ascending=False)

# En önemli ilk 10 geni görselleştirelim
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=features.head(10))
plt.title('Top 10 Important Features (Genes)')
plt.xlabel('Importance')
plt.ylabel('Gene')
plt.show()

# En önemli genleri yazdıralım
print("Top 10 Important Genes:")
print(features.head(10))
```
